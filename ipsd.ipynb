```python
# 2.12 Numpy & Pandas (pengantar)
import numpy as np
import pandas as pd

# Numpy array & operasi vektor
arr = np.array([1,2,3,4,5])
arr_mean = arr.mean()
arr2 = arr * 10

print("Array:", arr)
print("Rata-rata array:", arr_mean)
print("Array dikali 10:", arr2)
print()

# Pandas DataFrame sederhana
df = pd.DataFrame({
    "nama": ["Ani", "Budi", "Cici", "Dodi"],
    "nilai": [88, 75, 92, 70],
    "jurusan": ["TI", "SI", "TI", "SI"]
})

print("DataFrame:")
print(df)
print()

df_describe = df.describe(include=[int, float])
print("Deskripsi Statistik:")
print(df_describe)
```

```python
# 2.12 Numpy & Pandas (pengantar)
import numpy as np
import pandas as pd

# Numpy array & operasi vektor
arr = np.array([1,2,3,4,5])
arr_mean = arr.mean()
arr2 = arr * 10

print("Array:", arr)
print("Rata-rata array:", arr_mean)
print("Array dikali 10:", arr2)
print()

# Pandas DataFrame sederhana
df = pd.DataFrame({
    "nama": ["Ani", "Budi", "Cici", "Dodi"],
    "nilai": [88, 75, 92, 70],
    "jurusan": ["TI", "SI", "TI", "SI"]
})

print("DataFrame:")
print(df)
print()

df_describe = df.describe(include=[int, float])
print("Deskripsi Statistik:")
print(df_describe)
```

```python
# 2.13 Visualisasi dengan Matplotlib (pengantar)
import matplotlib.pyplot as plt

nilai = [88, 75, 92, 70]
nama = ["Ani", "Budi", "Cici", "Dodi"]

plt.figure()
plt.bar(nama, nilai)
plt.title("Nilai Mahasiswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()
```
